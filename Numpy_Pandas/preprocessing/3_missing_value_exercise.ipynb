```python
import pandas as pd
import numpy as np
```

```python
# Eaxmple from - https://chrisalbon.com/python/pandas_missing_data.html
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'],
            'age': [42, np.nan, 36, 24, 73],
            'sex': ['m', np.nan, 'f', 'm', 'f'],
            'preTestScore': [4, np.nan, np.nan, 2, 3],
            'postTestScore': [25, np.nan, np.nan, 62, 70]}

df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name',
                                        'age', 'sex', 'preTestScore', 'postTestScore'])

df
```

```python
df.isnull()
# 값이 없는(missing) value를 처리하기 위해 검색
```

```python
df.isnull().sum()
```

```python
df.isnull().sum() / len(df)
```

```python
# dropna() : Nan 값이 있는 데이터를 삭제
df_no_missing = df.dropna()
df_no_missing
```

```python
df
```

```python
df_cleaned = df.dropna(how = 'all')
# (how = 'all') : 전체가 NaN 값 일 경우에만 삭제
df_cleaned
```

```python
df['location'] = np.nan
df
```

```python
df.dropna(axis=1, how='all', inplace = True)
df
# 전부 다 NaN 인 column 삭제
```

```python
df
```

```python
df.dropna(axis=0, thresh=1) 
# thresh = 1 -> 데이터가 한개 미만이면 삭제한다.
```

```python
df.dropna(thresh = 5)
```

```python
df
```

```python
df.fillna(0)
# NaN 인 것들을 0으로 채워줌
# 단순히 0으로만 채워주는 게 아니라 평균값, 중간값등을 채워줄 수도 있음.
```

```python
df["preTestScore"].mean()
```

```python
df["preTestScore"].median()
```

```python
df["preTestScore"].mode()
```

```python
df["preTestScore"]
```

```python
df["preTestScore"].fillna(df["preTestScore"].mean(), inplace=True)
# preTestScore 의 NaN 값들을 평균값으로 채워준다.
```

```python
df
```

```python
df.groupby("sex")["postTestScore"].transform("mean")
```

```python
df["postTestScore"].fillna(
    df.groupby("sex")["postTestScore"].transform("mean"), inplace=True)

# postTestScore의 Nan 값들을
# 성별 기준으로 평균값으로 채워준다.
# Nan 값을 어떤 것으로 채울 지는 전문가의 지식이 필요하다!
```
