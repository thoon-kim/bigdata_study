** What are the top 5 most common jobs? **

```python
jobs = {}

for i in sal['JobTitle']:
    if i in jobs:
        jobs[i] += 1
    else:
        jobs[i] = 1
        
print(jobs)
```

```python
sorted(jobs.items(), key=lambda x:x[1], reverse=True)[:5]
```

```python
sal['JobTitle'].value_counts()[:5]
```

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
sal_2013 = sal[sal['Year'] == 2013]
jobs = {}

for i in sal_2013['JobTitle']:
    if i in jobs:
        jobs[i] += 1
    else:
        jobs[i] = 1
        
print(jobs)
```

```python
count = 0 
for i in jobs.values():
    if i == 1:
        count += 1
print(count)
```

```python
(sal[sal['Year'] == 2013]['JobTitle'].value_counts() == 1).values.sum()
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
count = 0

for i in sal['JobTitle']:
    if 'chief' in i.lower():
        count += 1
        
print(count)
```
