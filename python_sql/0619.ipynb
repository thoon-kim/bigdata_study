```python
# sakila db / payment table
# 월별 전체 매출, 평균 매출, 결제 횟수
# 위의 값을 출력하고
# 위의 각각의 값을 dataframe의 칼럼값으로 해서 출력
# DataFrame, SQL
```

```python
# DateFrame
sql = '''
    select *
    from payment;
'''

payment_df = pd.read_sql(sql, sakila_db)
```

```python
payment_df['month'] = payment_df['payment_date'].apply(lambda x: str(x)[:7])
payment_df = payment_df[['payment_id', 'amount', 'month']]
```

```python
# 평균 매출
payment_df.groupby('month')['amount'].mean()
```

```python
# 전체 매출
payment_df.groupby('month')['amount'].sum()
```

```python
# 결제 횟수
payment_df.groupby('month')['payment_id'].count()
```

```python
# payment_df.groupby('month')[['payment_id'].count(), ['amount'].sum(), ['amount'],mean()]

import numpy as np

payment_df.groupby('month').\
agg({'amount' : [np.mean, np.sum], 'payment_id':np.size})
## aggregation function
```

```python
# SQL
sql = '''
    select concat(year(payment_date), '/', month(payment_date)) as month, 
           sum(amount) as 전체_매출, avg(amount) as 평균_매출, count(payment_id) as 결제_횟수
    from payment
    group by month;
'''

sql = '''
    select left(payment_date, 7) as month, sum(amount) as 전체_매출, 
           avg(amount) as 평균_매출, count(payment_id) as 결제_횟수
    from payment
    group by month;
'''

pd.read_sql(sql, sakila_db)
```
