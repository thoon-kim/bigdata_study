```python
# 반복문으로 값을 받다가 엔터를 치면 종료하는 식으로
def fun():
    while 1:
        ipdata = input('문장을 입력하세요 : ')
        if ipdata == '':
            break
        f = open('myfile.txt', 'a')
        f.write(ipdata)
        f.close()
        
fun()
f = open('myfile.txt', 'r')
sen = f.read()
print(sen)
```

```python
# my_data에 저장 , 함수 파일을 넣으면 알파벳 몇개가 들어가있는지
def getTextFreq(filename):
    f = open(filename, 'r')
    data = f.read()
    ret = {}
    
    for c in data:
        if c >= 'A' and c <= 'z':
            if c in ret:
                ret[c] += 1
            else:
                ret[c] = 1
    return sorted(ret.items())

ret = getTextFreq('harry.txt')
print(ret)
```

```python
# 파일에서 단어의 갯수를 출력하는 함수를 만드세요
def countWord(filename, word):
    f = open(filename, 'r')
    data = f.readlines()
    count = 0
    for i in data:
        if i.find(word):
            count += 1
    return count

word = 'joanne'
ret = countWord('harry.txt', word)
print('[%s]의 갯수 : %d' %(word, ret))
```

```python
# 파일에서 단어의 갯수를 출력하는 함수를 만드세요

def countword(filename, word):
    count = 0
    f = open(filename, 'r')
    data = f.readlines() # readlines 로 읽어야만 한 줄씩 
    
    for i in data:
        count += i.count(word)
        print(i)
        
    f.close()
    
    return count
    
word = input('단어 : ')
ret = countword('subject.txt', word)
print('[%s] 는 %d 개'%(word, ret))
```
