### ban.txt 파일을 읽어서 딕셔너리로 출력

```python
D = {}
f = open('ban.txt', 'r')

for line in f:
    (key,val) = line.split() ## abc 111 로 나누어져있기 때문에 각각 할당
    D[key] = val
# #   D[line,split()[0]

D

```

```python
## ban_student.txt 파일을 읽어서 다음과 같이 출력

D = {}
f= open('ban_student.txt','r')
for line in f:
    key = line.split()[0]
    val = line.split()[1:]
    D[key] = val
    
D
```

```python
## stockcode.txt 파일을 read() 함수로 읽어서

D = {}
f= open('stockcode.txt','r')

a= f.read()
b= a.split()
i=0
while i<len(b):
    D[b[i]] = b[i+1]
    i +=2
    if i>len(b):
        break
D
```

```python
## stockcode.txt 파일을 readline() 함수로 읽어서

D= {}

f= open('stockcode.txt','r')

a = f.readline()
while a:
    key = a.split()[0]
    value =a.split()[1]
    D[key] =value
    a = f.readline()

D
```

```python
## stockcode.txt 에서 '삼성전자'의 거래소 번호 출력

D ={}
L = []
f = open('stockcode.txt','r')

a = f.readlines()
    
for i in a:
    (key,value)  = i.split()
    D[key] = value
    if(value == '삼성전자'):
        print(key)
        break
```

## 함수에 텍스트 입력받아 파일로 저장

```python
## 함수실행  -> 값 입력 -> myfile.txt 저장

def fun():
    ipdata = input('문장을 입력하세요 ')
    f = open('myfile.txt','w')
    f.write(ipdata)
    f.close()
    
fun()
f = open('myfile.txt','r')
lia = f.read()
print(lia)
```

## 반복문으로 값을 받다가 엔터를 치면 종료하는 식으로

```python
## 반복문으로 값을 받다가 엔터를 치면 종료하는 식으로

def fun():
    while 1:
        ipdata = input('문장을 입력하세요 ')
        if ipdata == '':
            break
        f = open('myfile.txt','a')
        f.write(ipdata)
        f.close()
fun()
f = open('myfile.txt','r')
lia = f.read()
print(lia)
```

```python
## 모범답안
## 사진 저장하고 복사하기

buf_size = 100 ### 100바이트씩 나누어서 복사하게만들어준다
f = open('세계지도.png','rb')
data = f.read(buf_size)

g = open('세계지도2.png','wb')

while data:
    data = f.read(buf_size)
    g.write(data)
    
f.close()
g.close()
```
