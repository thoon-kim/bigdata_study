```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples = 150,
                  n_features = 2,
                  centers = 3,
                  cluster_std = 0.5,
                  shuffle = True,
                  random_state = 0)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x[:, 0], x[:, 1],
            c = 'white', marker = 'o', edgecolor = 'black', s = 50)

plt.grid()
plt.tight_layout()

plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 3,
            init = 'random', ## default k-means++
            max_iter = 300,
            random_state = 0)

y_km = km.fit_predict(x)
```

```python
plt.scatter(x[y_km == 0, 0],
            x[y_km == 0, 1],
            s = 50, c = 'lightgreen',
            marker = 's', edgecolor = 'black',
            label = 'cluster 1')

plt.scatter(x[y_km == 1, 0],
            x[y_km == 1, 1],
            s = 50, c = 'orange',
            marker = 'o', edgecolor = 'black',
            label = 'cluster 2')

plt.scatter(x[y_km == 2, 0],
            x[y_km == 2, 1],
            s = 50, c = 'lightblue',
            marker = 'v', edgecolor = 'black',
            label = 'cluster 3')

plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s = 50, c = 'red',
            marker = '*', edgecolor = 'black',
            label = 'centroids')

plt.legend(scatterpoints = 1)
plt.grid()
plt.tight_layout()

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
ac = AgglomerativeClustering(n_clusters = 3,
                             affinity = 'euclidean',
                             linkage = 'complete')
y_ac = ac.fit_predict(x)

plt.scatter(x[y_ac == 0, 0],
            x[y_ac == 0, 1],
            s = 50, c = 'lightgreen',
            marker = 's', edgecolor = 'black',
            label = 'cluster 1')

plt.scatter(x[y_ac == 1, 0],
            x[y_ac == 1, 1],
            s = 50, c = 'orange',
            marker = 'o', edgecolor = 'black',
            label = 'cluster 2')

plt.scatter(x[y_ac == 2, 0],
            x[y_ac == 2, 1],
            s = 50, c = 'lightblue',
            marker = 'v', edgecolor = 'black',
            label = 'cluster 3')

plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s = 50, c = 'red',
            marker = '*', edgecolor = 'black',
            label = 'centroids')

plt.legend(scatterpoints = 1)
plt.grid()
plt.tight_layout()

plt.show()
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps = 0.6, min_samples = 3, metric = 'euclidean') # eps  / min_samples    
y_db = db.fit_predict(x)

plt.scatter(x[y_db == 0, 0],
            x[y_db == 0, 1],
            s = 50, c = 'lightgreen',
            marker = 's', edgecolor = 'black',
            label = 'cluster 1')

plt.scatter(x[y_db == 1, 0],
            x[y_db == 1, 1],
            s = 50, c = 'orange',
            marker = 'o', edgecolor = 'black',
            label = 'cluster 2')

plt.scatter(x[y_db == 2, 0],
            x[y_db == 2, 1],
            s = 50, c = 'lightblue',
            marker = 'v', edgecolor = 'black',
            label = 'cluster 3')

plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s = 50, c = 'red',
            marker = '*', edgecolor = 'black',
            label = 'centroids')

plt.legend()
plt.tight_layout()

plt.show()
```
