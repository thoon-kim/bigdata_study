```python
# 지자체 코드

import urllib
import json
from pprint import pprint

ServiceKey = "86cs3omKM4YBLXw30AHy5IsdlM%2Fsf0MAU2tXn8LYd3uWyZQyryjew1SQgckNLc2lkYM%2B26W9W3ue4chSd2hI9A%3D%3D"
# url = "http://apis.data.go.kr/B552584/RfidFoodWasteService/getCityList?ServiceKey={}&type=json".format(ServiceKey)
district_code = list()

for i in range(1, 20):
    url = "http://apis.data.go.kr/B552584/RfidFoodWasteService/getCityList?ServiceKey={}&type=json&page={}".format(ServiceKey, i)
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    
    count = 0
    
    if(rescode==200):
        response_body = response.read()
        dic = json.loads(response_body.decode('utf-8'))
        if dic['data']['list']:
            for j in range(len(dic['data']['list'])):
                if count <= dic['data']['count']:
                    district_code.append(dic['data']['list'][j])
                    count += 1
                else:
                    break
    else:
        break
        print("Error Code:" + rescode)
        
for i in range(len(district_code)):
    print(district_code[i])
```
