```python
import pandas as pd
import numpy as np

df7 = pd.DataFrame(np.random.randn(5, 3),
                  index = ['a', 'c', 'e', 'f', 'h'],
                  columns = ['one', 'two', 'three'])
df8 = df7.reindex(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])
df8
```

```python
# NaN => Not a Number
# Check isnull(), notnull()
df8['one'].isnull()
df8['one'].notnull()
df8.isnull()
df8.notnull()
df8.fillna(0) # Nan값인 데이터를 0으로
df8.sum()
df8['one'].sum()
df8.dropna() # NaN 값인 데이터행을 삭제

# groupby
# groupby('key')
# groupby(['key1', 'key2'])
# groupby(axis = 1)
# groupby('key').groups
```

```python
# Transforamtion

data = {
    'Team':['A', 'B', 'C', 'A', 'B', 'C', 'A', 'A', 'B', 'C', 'A', 'A'],
    'Rank':[1,2,2,3,4,2,1,2,2,3,4,2],
    'Year':['2014', '2015', '2014', '2015', '2016', '2017', 
            '2016', '2014', '2015', '2017', '2014', '2015']
}

df10 = pd.DataFrame(data)
df10

grp = df10.groupby('Team')
score = lambda x : (x - x.mean())
grp.transform(score)
```

```python
import pandas as pd

left = pd.DataFrame({
    'id' : [1,2,3,4,5],
    'Name' : ['Tom', 'Amy', 'John', 'Young', 'Andrew'],
    'subject_id' : ['sub1', 'sub2', 'sub3', 'sub1', 'sub2']
})

right = pd.DataFrame({
    'id' : [1,2,3,4,5],
    'Name' : ['Tom2', 'Amy2', 'John2', 'Young2', 'Andrew2'],
    'subject_id' : ['sub1', 'sub1', 'sub4', 'sub3', 'sub2']
})

pd.merge(left, right, on='id')
pd.merge(left, right, on=['id', 'Name'])
pd.merge(left, right, on=['id', 'subject_id'])
```
