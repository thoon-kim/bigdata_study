```python
# 감정점수를 계산하는 함수 작성

def emotionf(s1):
    score = 0
    pos = ('positive', 'anticipation','trust','joy','surprise')
        
    for i in s1:
        if i in pos:
            score += 1
        else:
            print(i)
            score -= 1
                
    return score
```

```python
# 감정점수를 계산하는 함수 작성

def emotionf(*s1):
    score = 0
    length = len(s1[0])
    pos = ('positive', 'anticipation','trust','joy','surprise')
    
    for i in range(length):
        if s1[0][i] in pos:
            score += 1
        else:
            score -= 1
                
    return score
```

```python
emotionf(('positive', 'anticipation','trust','joy','surprise'))
```

```python
import pandas as pd
import glob
from afinn import Afinn
from nltk.corpus import stopwords 
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
pos_review=(glob.glob("data\\aclimdb\\train\\pos\\*.txt"))[:20]
```

```python
# f = open(pos_review, 'r')
# pos_review = f.readlines()[0]
# f.close()
```

```python
lines_pos=[]
for i in pos_review:
    try:
        f = open(i, 'r')
        temp = f.readlines()[0]
        lines_pos.append(temp)
        f.close()
    except Exception as e:
        continue
```

```python
print(lines_pos)
```

```python
len(lines_pos)
```

```python
# f = open(pos_review, 'r')
# pos_lines = f.readlines()[0]
# f.close()
```

```python
neg_review=(glob.glob("data\\aclimdb\\train\\neg\\*.txt"))[:20]
```

```python
# f = open(neg_review, 'r')
# neg_review = f.readlines()[0]
# f.close()
```

```python
lines_neg=[]
for i in neg_review:
    try:
        f = open(i, 'r')
        temp = f.readlines()[0]
        lines_neg.append(temp)
        f.close()
    except Exception as e:
        continue
```

```python
len(lines_neg)
```

```python
total_text = lines_neg + lines_pos
```

```python
total_text
```

```python
len(total_text)
```

```python
total_text[25]
```

```python
# x = np.array(["pos", "neg"])
# class_Index = np.repeat(x, [len(lines_pos), len(lines_neg)], axis=0)
```

```python
neg_index = np.repeat('neg', 20)
neg_index = list(neg_index)
```

```python
pos_index = np.repeat('pos', 20)
pos_index = list(pos_index)
```

```python
class_index = neg_index+pos_index
```

```python
print(class_index)
```

```python
stop_words = stopwords.words('english')
```

```python
vect = TfidfVectorizer(stop_words=stop_words).fit(total_text)
# Tfidf ----> TF/DF // 특정 단어가 특정 문서에서 차지하는 중요도
# TF 특정단어가 하나의 문서에서 출현한 빈도
# DF 특정단어가 출현한 문서 빈도
```

```python
# df(document frequency) - 전체 문서에서 특정 단어를 포함하는 문서의 빈도
# tf(term frequency) - 특정 단어가 하나의 문서에서 나타난 빈도
# tfidf - tf/df
```

```python
X_train_vectorized = vect.transform(total_text)

# X_train_vectorized.index = class_Index
import pandas as pd

pd.DataFrame(X_train_vectorized.toarray(), columns = vect.get_feature_names())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train_vectorized, class_index)
```
